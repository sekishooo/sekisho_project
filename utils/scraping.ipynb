```python
i = 1
while browser.find_elements_by_xpath('//*[@id="btn-more"]') != []:
    print(i,end=" ")
    browser.find_element_by_xpath('//*[@id="btn-more"]').click()
    sleep(5)  
    i = i+1
```

```python
links = browser.find_elements_by_class_name("massonary-title-mobile")
#print(links[0])

url = [0]*10000
i=0
for link in links:
    link.get_attribute('href')
    url[i] = link.get_attribute('href')
    #print(i," ",url[i])
    i = i+1
```

```python
#一括収集するやつ


group_num=0
#2509/2520
#1599/1896
#3055/3388
#1485/1962
#135/224
#3105/5099
#1013/1255

i=0

for link in links:
    group_num = int(i/500)
    p = 'C:/Users/関祥真/Pictures/wikiArt/all_pictures/'
    
    #if i==510:
    #    break
    print(i,end=" ")
    browser.get(url[i])
    sleep(2)#とりあえず3でやってみる
    path = '//*[@id="image-link"]/div/img'
    imageurl = browser.find_element_by_xpath(path).get_attribute("src")
    def download(url, path):
            #print(url)
            request = urllib2.urlopen(url, context=context)
            #print(url)
            filename = url.split('/')[-1]
            #print(filename)
            saveData = open(path + filename, 'wb')
            saveData.write(request.read())
            saveData.close()
            print('>>> get:' + filename)
            request.close()
    i = i+1
    download(imageurl, p)
```
